```python
#staticmethod decorator
class HCl:
  name = "Dharmesh"
  language = "Python"
  salary =45000000

  def getinfo(self):
   print(f"the language is {self.language} and {self.salary}" )

  def greet(Self):
    print("Welcome")

  @staticmethod
  def myfunc():
    print("Self declared")

rahul = HCl()
rahul.language = "Java"
rahul.getinfo()
rahul.greet()
rahul.myfunc()
```

```python
#dunder method == __init__(self)

class HCl:
  name = "Dharmesh"
  language = "Python"
  salary =45000000

  def __init__(self):
    print("this does not required function calling")

  def getinfo(self):
   print(f"the language is {self.language} and {self.salary}" )

  def greet(Self):
    print("Welcome")

  @staticmethod
  def myfunc():
    print("Self declared")

rahul = HCl()
rahul.language = "Java"
rahul.getinfo()
rahul.greet()
rahul.myfunc()
```

```python
class HCl:
  name = "Dharmesh"
  language = "Python"
  salary =45000000

  def __init__(self,language,salary, name ):
    self.name= name
    self.salary = salary
    self.language = language
    print("this does not required function calling")

  def getinfo(self):
   print(f"the language is {self.language} and {self.salary}" )

  def greet(Self):
    print("Welcome")

  @staticmethod
  def myfunc():
    print("Self declared")


rahul = HCl("Rahul","150000","PHP")
print(rahul.name,rahul.salary,rahul.language)
rahul.getinfo()
rahul.greet()
rahul.myfunc()
```
