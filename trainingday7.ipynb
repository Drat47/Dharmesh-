```python
from collections import deque
q = deque()
q.append(1)
q.append(2)
q.append(3)
print(q)
print("\n elements dequed" )
print(q.popleft())
print(q.popleft())

print(q)

```

```python
#usig queue module
from queue import LifoQueue
stack = LifoQueue(maxsize = 3)
print(stack.qsize())
stack.put(1)
stack.put(2)
stack.put(3)
print(stack.full())
print(stack.qsize())
```

```python
#get method functionto pop

# get() function to pop
# element from stack in
# LIFO order
print('\nElements popped from the stack')
print(stack.get())
print(stack.get())
print(stack.get())

print("\nEmpty: ",stack.empty())
```

```python
#crweate an algorithm find the lowest value in the list
#ways to replace a element in the list in many ways
my_array = [12,34,54,65,33]
minval = my_array[0]

for i in my_array:
    if i < minval:
        minval = i
print(minval)
```

```python
fruits = ['apple','banana', 'cherry']
newlst = []
for x in fruits:
    if 'a' in x:
        newlst.append(x)
print(newlst)
```

```python
#list comprehension
fruits = ['apple','banana', 'cherry']
newlst = [x for x in fruits if 'a' in x]
print(newlst)

#ITERABKES IN ONE LINE
newls = [x for x in range(10)]
print(newls)

#EXPRESSION
newls = [x.upper() for x in fruits if x!= "Kiwi"]
print(newls)

```
